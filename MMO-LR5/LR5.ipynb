```python
import spacy
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from spacy import displacy
import warnings
warnings.filterwarnings('ignore')
```

```python
with open('text.txt', 'r') as fp:
    text = fp.read()
text
```

<h3>Токенизация</h3>

```python
nlp = spacy.load('en_core_web_sm')
spacy_text = nlp(text)
for t in spacy_text[:10]:
    print(t)
print('...')
```

<h3>Частичная разметка</h3>

```python
for token in spacy_text[:10]:
    print('{} - {} - {}'.format(token.text, token.pos_, token.dep_))
print('...')
```

<h3>Лемматизация</h3>

```python
for token in spacy_text[:10]:
      print(token, token.lemma, token.lemma_)
```

<h3>Выделение (распознавание) именованных сущностей</h3>

```python
for ent in spacy_text.ents:
    print(ent.text, '->', ent.label_)
```

```python

displacy.render(spacy_text, style='ent', jupyter=True)
```

<h3>Разбор предложения</h3>

```python
displacy.render(spacy_text, style='dep', jupyter=True)
```
